```python
import music21 as m21
from itertools import product
import random
```

```python
whole_notes_no_octave = ["C", "D", "E", "F", "G", "A", "B"]
octaves = [2, 3, 4, 5]
whole_notes = [n+str(o) for o, n in product(octaves, whole_notes_no_octave)]
whole_notes = whole_notes[whole_notes.index("E2"):]
print(whole_notes)
```

```python
clef = m21.clef.TrebleClef()
s = m21.stream.Stream()#[clef])
for note_name in whole_notes:
    note = m21.note.Note(note_name)
    note.duration = m21.duration.Duration("whole")
    s.append(note)
s.show()
```

```python
# Random notes
note_name = random.choice(whole_notes)
s = m21.stream.Stream()#[clef])
note = m21.note.Note(note_name)
note.duration = m21.duration.Duration("whole")
s.append(note)
s.show()
```

```python
t = m21.stream.Part()
b = m21.stream.Part()
responses = []
for _ in range(10):
    note_name = random.choice(whole_notes)
    responses.append(note_name)
    note = m21.note.Note(note_name)
    note.duration = m21.duration.Duration("whole")
    r = m21.note.Rest()
    r.duration = m21.duration.Duration("whole")
    if whole_notes.index(note_name) < whole_notes.index('C4'):
        b.append(note)
        t.append(r)
    else:
        b.append(r)
        t.append(note)
s = m21.stream.Score()
s.insert(0, t)
s.insert(0, b)
s.show()
```
